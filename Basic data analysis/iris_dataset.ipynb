```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
iris = pd.read_csv('IRIS.csv')
```

```python
print("First 5 rows of dataset:")
print(iris.head())
```

```python
# Basic info about the dataset
print("Dataset Info:")
print(iris.info())

print("\nShape of dataset:", iris.shape)  # rows and columns
print("\nColumn Names:", iris.columns.tolist())

print("\nData Types:")
print(iris.dtypes)

print("\nMissing Values in Each Column:")
print(iris.isnull().sum())

# ---- 2. Descriptive Statistics ----
print("\nSummary Statistics:")
print(iris.describe())

# Value counts for species
print("\nSpecies Distribution:")
print(iris["species"].value_counts())
```

```python
# ---- Basic Data Analysis ----
# Calculate average of sepal_length
avg_sepal_length = iris['sepal_length'].mean()
print("\nAverage Sepal Length:", avg_sepal_length)

# Average by species
avg_by_species = iris.groupby("species")["sepal_length"].mean()
print("\nAverage Sepal Length by Species:")
print(avg_by_species)
```

```python
# ---- Visualization ----

# 1. Bar chart - average sepal length by species
avg_by_species.plot(kind="bar", color=["skyblue", "orange", "green"])
plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.show()

# 2. Scatter plot - Sepal length vs Sepal width
plt.scatter(iris["sepal_length"], iris["sepal_width"], alpha=0.7, c="blue")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

# 3. Heatmap - correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()


# ---- Insights ----
print("\nInsights:")
print("1. Average sepal length is around", round(avg_sepal_length, 2))
print("2. Virginica species has the largest average sepal length.")
print("3. Scatter plot shows a positive correlation between sepal length and width.")
print("4. Heatmap shows strong correlation between petal length & petal width.")
```
